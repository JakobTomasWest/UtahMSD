### Nested Query as Condition
> x is a value and A is a multi-set
1. 
```sql
x in A
```
- True if x in A
2. 
```sql
exists A
```
- True if A is not empty
3. 
```sql
-- <op> operation: ==, !=, >=, >, <, <=
x <op> any A
```
- True if there exists a y in A such that x <op> y is true
4. 
```sql
x <op> all A
```
- True if for all y in A, x <op> y is true
#### Example
![](<images/Nested Query as condition 1.png>)
```sql
select * from students 
where sID not in (
  select sID from student 
    natural join enrolled 
      natural join Courses 
      where Name = 'Databases')
```

![](<images/Nested Query as condition 2.png>)
```sql
select Student.Name from Student
where Student.DOB > all (select DOM from Student 
  natural join Enrolled 
    natural join Courses
    where Courses.Name = 'Database')
```
![](<images/Nested Query as condition 3.png>)
```sql
select s.Name
from Students s
where not exists (
  select c.cID
  from Courses c
  where not exists(
    select e.cID
    from Enrolled e
    where e.cID = c.cID
    and e.sID = s.sID
  )
);
```
- Separate into 3 parts
  - If the given sID and cID combination can't be found in the table=> return empty
```sql
select e.cID
    from Enrolled e
    where e.cID = c.cID
    and e.sID = s.sID
```
  - If the selection above return empty -> not exists: condition satisfied => will return the cID which didn't fulfill the inner selection
```sql
select c.cID
  from Courses c
  where not exists(
    select e.cID
    from Enrolled e
    where e.cID = c.cID
    and e.sID = s.sID
  )
```
  - If the 2nd inner selection returns a cID -> not exists: dissatisfied => will not select that s.Name
```sql
select s.Name
from Students s
where not exists (
  select c.cID
  from Courses c
  where not exists(
    select e.cID
    from Enrolled e
    where e.cID = c.cID
    and e.sID = s.sID
  )
);
```

## ACID Transaction
> ACID: Atomicity Consistency Isolation Durability
- Atomicity : 100-0 
- Consistency: never result in violated constraints
- Isolation: don't interfere with each other(sometimes need to synchronization)
```sql
-- show isolation level
select @@transaction_isolation;
-- change isolation level
set session transaction isolation level serializable;
```
  - read uncommitted (dirty read)
    - can read uncommitted data from other sessions
    - might read nonexistent data
  - read committed (phantom read)
    - can read committed data from other sessions
    - we get different result if the data hs been changed between 2 read commands
  - repeatable read(default)
    - each transaction(session) has it's own snapshot of the DB 
    - consistency of the read results
  - serializable
    - we will need to wait to see the result until the another session to commit (like mutex but not actually locking the tables)
    - the reason we do this is because we might modify the table=> we need the info in the table
- Durability: won't loss operation
### Atomicity & Consistency

        - the last element points to it's closest sibling -> like a linked list
