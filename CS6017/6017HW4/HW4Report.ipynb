### Step 3: Model Training

Four models were trained using scikit-learn:

Logistic Regression

Pipeline: StandardScaler + LogisticRegression(class_weight='balanced')

Accuracy: ~17%

Macro F1: ~0.06

Observations: Performed poorly overall due to linear assumptions not matching poker hand complexity. The use of class weights caused it to over-predict rare classes like Full House, resulting in poor overall class-wise balance. However, not using the class weights still yielded poor results. 

Decision Tree

Config: DecisionTreeClassifier(max_depth=10)

Accuracy: ~53.5%

Macro F1: ~0.12

Weighted F1: ~0.52

Observations: Best performing model by accuracy. Captured rule-like patterns well and handled majority classes effectively, though rare classes remained underrepresented. It seems difficult to represent such imbalanced anomolies. 

K-Nearest Neighbors (KNN)

Pipeline: StandardScaler + KNeighborsClassifier(k=8, weights='distance')

Accuracy: ~50.9%

Macro F1: ~0.11

Weighted F1: ~0.49

Observations: Performed similarly to the tree; limited performance on rare classes due to few neighbors.

Support Vector Machine (SVM)

Pipeline: StandardScaler + SVC(kernel='rbf', C=5, class_weight='balanced')

Accuracy: ~36.7%

Macro F1: ~0.11

Weighted F1: ~0.37

Observations: Captured nonlinear decision boundaries better than Logistic Regression but struggled with both performance and training speed. Still weak on rare classes even with class weighting. ### Step 5: Reflection

Decision Trees performed best overall in this classification task, offering a good balance of accuracy and interpretability. KNN performed similarly on this small dataset but would be inefficient on larger datasets due to its lazy prediction approach. Logistic Regression struggled with the nonlinear nature of poker hand chances and overcompensated for rare classes. SVM handled nonlinearity better but still missed many rare cases and was slower to train.

The biggest challenge was extreme class imbalance, which most models struggled to address. class_weight='balanced' helped slightly but led to false positives in some rare categories. Future improvements could include using ensemble methods, SMOTE.
