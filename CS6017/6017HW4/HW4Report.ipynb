Decision Tree had the best overall accuracy and interpretability. KNN performed similarly but would not scale well with more data. SVM captured some nonlinearity but was computationally expensive and still biased. Logistic Regression was the weakest due to its linear nature.

### Step 5: Reflection

Decision Trees performed best overall in this classification task, offering a good balance of accuracy and interpretability. KNN performed similarly on this small dataset but would be inefficient on larger datasets due to its lazy prediction approach. Logistic Regression struggled with the nonlinear, rule-based nature of poker hands and overcompensated for rare classes. SVM handled nonlinearity better but still missed many rare cases and was slower to train.

The biggest challenge was extreme class imbalance, which most models struggled to address. class_weight='balanced' helped slightly but led to false positives in some rare categories. Future improvements could include using ensemble methods, SMOTE, or a larger dataset with better class balance.
